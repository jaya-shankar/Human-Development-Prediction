```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv("/content/wcde_data (1).csv")   
df = pd.DataFrame(data)
```

```python
edu_level = { 'In_Primary_OL'     : [ 'No Education', 'Incomplete Primary'], 
              'Primary_OL'        : [ 'No Education', 'Incomplete Primary','Primary'],
              'Lower_Secondary_OL': ['No Education', 'Incomplete Primary','Primary', 'Lower Secondary'],
            }
```

```python
age_groups = list(df.Age.unique())
age_groups
```

```python
df_list = []

for grp in age_groups:
  df_list.append(df[df.Age == grp])
```

```python
def change_yr(age):
  a = int(age[:2])
  if a==10:
    return str(1955-(100-20))
  else:
    return str(1955-(a-20))

for i in range(len(df_list)):
  df_list[i]['Year'] = change_yr(age_groups[i])
  # df_list[i]['Age'] = '20--24'

df_list
```

```python
df_list[0]
```

```python
for i in range(1,len(df_list)):
  df_list[0] = df_list[0].append(df_list[i], ignore_index=True)
```

```python
df = df_list[0]
```

```python
for i in edu_level:
  sr = df.loc[df['Education'].isin(edu_level[i])]
  sr.to_csv("/content/"+i+'_1955.csv', encoding='utf-8', index=False)
```

```python
def init_timeline():
  timeline = [i for i in range(1875,1956)]
  timeline_dic={c:{t : 0 if not t%5 else np.NaN for t in timeline  } for c in countries}
  return timeline_dic
```

```python
def convert_to_df():
  dft = pd.DataFrame.from_dict(timeline_dic,orient='index')
  dft.insert(0, "Country", countries, True)
  return dft
```

```python
for e in edu_level:
  timeline_dic = init_timeline()
  data = pd.read_csv("/content/"+e+"_1955.csv")   
  dfs = pd.DataFrame(data)
  for j in range(len(data)):
    timeline_dic[dfs.iloc[j]['Area']][dfs.iloc[j]['Year']] += dfs.iloc[j]['Distribution']
    round(timeline_dic[dfs.iloc[j]['Area']][dfs.iloc[j]['Year']], 2)
  dft = convert_to_df()
  dft.to_csv("/content/"+e+'_1955.csv', encoding='utf-8', index=False)
```

```python
years = [(y,y+5) for y in range(1875,1951,5)]
datasets = [ e for e in edu_level]

for name in datasets:
  data = pd.read_csv("/content/"+name+'_1955.csv')   
  df = pd.DataFrame(data)
  for r in range(len(df)):
    for s,e in years:
      df.at[r,str(s)] = round(df.iloc[r][str(s)],2)
      df.at[r,str(e)] = round(df.iloc[r][str(e)],2)
      d   = df.iloc[r][str(e)] - df.iloc[r][str(s)]
      inc = d/5
      i=1
      for y in range(s+1,e):
        df.at[r,str(y)] = round(df.iloc[r][str(s)] + inc*i,2)
        i+=1
      for yr in range(1956,1960):
        df[str(yr)] = np.NaN

  df.to_csv("/content/"+name+'_1955.csv', encoding='utf-8', index=False)
```

## Combine Datasets



```python
import pandas as pd
import numpy as np
import csv
```

```python
for edu in edu_level:
  df1 = pd.read_csv("/content/"+edu+".csv")
  df2 = pd.read_csv("/content/"+edu+"_1955.csv")

  countries1 = set((df1.Country.unique()))
  countries2 = set((df2.Country.unique()))

  missing_countries = list(countries1 - countries2)
  df1.set_index(["Country"], inplace=True)
  df2.set_index(["Country"], inplace=True)
  for c in missing_countries:
    try:
      df1.drop(index = c, inplace=True)
    except:
      pass
  # df3 = pd.concat([df2, df1], axis=1)
  df1.reset_index(inplace = True)
  df2.reset_index(inplace = True)
  df3 = pd.merge(df2, df1, left_index=True, right_index=True)
  df3.drop(columns=['Country_y'],inplace=True)
  for r in range(len(df3)):
      s = 1955
      e = 1960
      df3.at[r,str(s)] = round(df3.iloc[r][str(s)],2)
      df3.at[r,str(e)] = round(df3.iloc[r][str(e)],2)
      d   = df3.iloc[r][str(e)] - df3.iloc[r][str(s)]
      inc = d/5
      i=1
      for y in range(s+1,e):
        df3.at[r,str(y)] = round(df3.iloc[r][str(s)] + inc*i,2)
        i+=1
  df3.rename(columns={"Country_x": "Country"},inplace=True)
  df3.to_csv('/content/'+edu+'_complete.csv', encoding='utf-8', index=False)
```
